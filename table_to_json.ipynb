```python
import pandas as pd
import json
```

# Full dataset

```python
nodes = pd.read_excel('BASE DE DATOS MAPA ACTORES TRACE-HARMONIZE.xlsx', sheet_name='Elements')
nodes
```

```python
nodes.index.name = 'id'
nodes.columns.values
column_short_names = [
    'name',
    'vis',
    'actor_type',
    'entity_type',
    'sector_type',
    'org_type',
    'scale',
    'rel_health_sys',
    'gen_user_type',
    'act_user_type',
    'rel_sivigila',
    'tags',
    'exp_area',
    'rel_etvs',
    'problematization',
    'poss_interest',
    'description',
    'website',
    'twitter',
    'notes'
]
nodes.columns = column_short_names
nodes.reset_index(inplace=True)
nodes.fillna('', inplace=True)
nodes
```

```python
nodes_json = [{'data': i} for i in nodes.to_dict(orient='records')]
nodes_json
```

```python
name_id_mapper = nodes.loc[:,['id', 'name']].set_index('name').to_dict()['id']
name_id_mapper
```

```python
edges = pd.read_excel('BASE DE DATOS MAPA ACTORES TRACE-HARMONIZE.xlsx', sheet_name='Connections')
edges.columns = ['source', 'target']
edges['source'] = edges['source'].map(name_id_mapper)
edges['target'] = edges['target'].map(name_id_mapper)
edges.dropna(inplace=True)
edges['source'] = edges['source'].astype(int)
edges['target'] = edges['target'].astype(int)
edges_json = [{'data': i} for i in edges.to_dict(orient='records')]
edges
```

```python
edges_json = [{'data': i} for i in edges.to_dict(orient='records')]
edges_json
```

```python
with open('actors.json', 'w') as json_file:
    json.dump(nodes_json + edges_json, json_file, indent=4)
```
